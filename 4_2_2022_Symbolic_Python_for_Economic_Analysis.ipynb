* we have 3 constant (a,b,c) and 2 parameters $q_1$ and $q_2$
* creating symbols for this five symbols 

```python
a,b,c,q1,q2 = symbols('a b c q1 q2')
a,b,c,q1,q2
```

```python
print('Inverse demand curve:')
p = a-b*(q1+q2)
p
```

```python
print('Total cost for firm 1:')
tc1 = c*q1
tc1
```

```python
print('Total cost for firm 2:')
tc2 = c*q2
tc2
```

```python
print('Total profit function for firm 1:')
profit1 = p*q1 - c*q1
profit1
```

```python
print('Expanded expression of Total profit function for firm 1:')
print('Total profit function for firm 1:')
profit1 = p*q1 - c*q1
expand(profit1)
```

```python
print('Total profit function for firm 2:')
profit2 = p*q2 - c*q2
profit2
```

```python
print('Expanded expression of Total profit function for firm 2:')
profit2 = p*q2 - c*q2
expand(profit2)
```

# First order condition
* differentiate each profit function with respect to the firm's quantity  
* in symbolic math, using **diff** as follow:
* diff(euqation, variable) 
* in our setting diff(profit function for firm a, firm a's quantity)
* answer for differenciation

# Firm 1 first order condition + best response function

```python
print('First order condition for firm 1:')
foc_q1 = diff(profit1, q1)
foc_q1 = Eq(foc_q1)
foc_q1 
```

```python
print('Best response function for firm 1:')
BRF1 = solve(foc_q1, q1)
BRF1
```

```python
BRF1 = (a - b*q2 - c)/(2*b)
expand(BRF1)
```

# Firm 2 first order condition + best response function

```python
print('First order condition for firm 2:')
foc_q2 = diff(profit2, q2)
foc_q2
```

```python
print('Best response function for firm 1:')
BRF2 = solve(foc_q2, q2)
BRF2
```

```python
BRF2 =  (a - b*q1 - c)/(2*b)
expand(BRF2)
```

# Optimal quantity for firm 1
* solve function give you a solution for 
* equation = 0, the solution is based on the definend variable 
* x+3y=0, if you want to solve it by x, then solve(x+3y, x), it will give you x=-3y

```python
# plug BRF2 into BRF1 then you can get the optimal quantity for firm 1
print('optimal qunaitity for firm 1:')
optimal_q1 = solve((a-b*BRF2-c)/(2*b)-q1,q1)
optimal_q1
```

```python
print('optimal qunaitity for firm 1:')
optimal_q1 =(a - c)/(3*b)
optimal_q1
```

# Optimal quantity for firm 2

```python
# plug optimal qunaity into BRF2 then you can get the optimal quantity for firm 2
print('optimal qunaitity for firm 2:')
optimal_q2 = solve((a-b*optimal_q1-c)/(2*b)-q2,q2)
optimal_q2
```

```python
print('optimal qunaitity for firm 2:')
optimal_q2 = (a - c)/(3*b)
optimal_q2
```

# Optimal price
* plug optimal quantity 1 and 2 into inverse demand function
* $ p^*=a-b(q^*_1+q^*_2)$

```python
print('optimal price for firms 1 and 2:')
optimal_price = a-b*(optimal_q1+optimal_q2)
optimal_price
```

* **factor** function express equation more simple format : opposite effect with **expand** function

```python
print('optimal price for firms 1 and 2:')
factor(optimal_price)
```

# Maximum Profit for firm 1
* plug optimal price and quantity for firm 1 into profit fucntion for firm 1

```python
print('maximum profit for firm 1:')
max_profit1 = optimal_price*optimal_q1 -c*optimal_q1
max_profit1 
```

```python
print('maximum profit for firm 1:')
factor(max_profit1 )
```

# Maximum Profit for firm 2
* plug optimal price and quantity for firm 2 into profit fucntion for firm 2



```python
print('maximum profit for firm 2:')
max_profit2 = optimal_price*optimal_q2 -c*optimal_q2
max_profit2 
```

```python
print('maximum profit for firm 1:')
factor(max_profit2)
```
